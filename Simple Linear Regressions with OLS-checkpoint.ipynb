Those look pretty good, let's start building our OLS lr model!

```python
X = final_df.drop(columns='saleprice')
y = final_df['saleprice']
```

```python
lrm = linear_model.LinearRegression()

lrm.fit(X, y)

print('Coefficients: \n', lrm.coef_)
print('\nIntercept: \n', lrm.intercept_)
```

Ok, model built! I'm sure we'll learn how to evaluate it soon...

## Gauss assumptions:

### 1) Linearity of features and target

```python
predictions = lrm.predict(X)
```

```python
for column in X:
    sns.scatterplot(X[column], predictions)
    plt.show()
```

### Assumption 2: Error near 0

```python
errors = y - predictions
print('Mean of our erros is {}'.format(errors.mean()))
```

Ok, this looks good!

### Assumption 3: homoscedasticity

```python
sns.scatterplot(predictions, errors)
plt.title('Residual vs Predicted')
plt.xlabel('Predicted')
plt.ylabel('Residual')
plt.axhline(y=0)
plt.show()
```

So, we are underestimating really large values. In fact, it seems our errors grow larger around $400,000

```python
from scipy.stats import bartlett
from scipy.stats import levene

bart_stats = bartlett(predictions, errors)
lev_stats = levene(predictions, errors)

print("Bartlett test statistic value is {0:3g} and p value is {1:.3g}".format(bart_stats[0], bart_stats[1]))
print("Levene test statistic value is {0:3g} and p value is {1:.3g}".format(lev_stats[0], lev_stats[1]))
```

Well, these are very low p_values. So, it seems that our data is heteroscedastic. :(

This is probably due to outliers!

### Assumption 4: Low Multicollinearity

I would assume our heatmap from before should work. Let's check it using our 'X' variable.

```python
sns.heatmap(X.corr(), annot=True, linewidth=0.5)
plt.show()
```

It looks like totrmsabvgrd and grlivarea are strongly correlated. This makes sense. We should probably remove these.

### Assumption 5: uncorrelated error terms

```python
plt.plot(errors)
plt.show()
```

These don't look too correlated. So that's good!

```python
sns.distplot(errors)
plt.show()
```

```python
rand_nums = np.random.normal(np.mean(errors), np.std(errors), len(errors))


sns.scatterplot(np.sort(rand_nums), np.sort(errors)) # we sort the arrays
plt.xlabel("the normally distributed random variable")
plt.ylabel("errors of the model")
plt.title("QQ plot")
plt.show()
```

Ok, not too bad. Some strange happenings at the tail ends of our qq plot. But otherwise ok!

### Assumption 6: Features uncorrelated from the errors

```python
X_errors = pd.concat([X, errors], axis=1)
```

```python
sns.heatmap(X_errors.corr(), annot=True, linewidth=0.5)
plt.show()
```

Well, there are some pretty strong correlations here!

```python
from scipy.stats import jarque_bera
from scipy.stats import normaltest
```

```python
jb_stats = jarque_bera(errors)
norm_stats = normaltest(errors)

print("Jarque-Bera test statistics is {0} and p value is {1}".format(jb_stats[0], jb_stats[1]))
print("Normality test statistics is {0} and p value is {1}".format(norm_stats[0], norm_stats[1]))
```

So, our errors are not-normally distributed, although they slightly looked like it

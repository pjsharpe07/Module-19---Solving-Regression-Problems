```python
engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
df = pd.read_sql_query('select * from weatherinszeged',con=engine)

# no need for an open connection, as we're only doing a single query
engine.dispose()


df.head()
```

```python
#allow us to see all columns and rows w/out truncation
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
X = df[['humidity', 'windspeed', 'windbearing']]
y = df['temperature']
```

```python
lrm  = linear_model.LinearRegression()

lrm.fit(X, y)

print('Coefficients: \n', lrm.coef_)
print('\nIntercept: \n', lrm.intercept_)
```

### Assumption 1: Linearity of features and target

```python
predictions = lrm.predict(X)
```

Not really sure if there is a linear relationship between our features and target here...

### Assumption 2: Error near 0

```python
errors = predictions - y
print('Mean of our errors is {}'.format(errors.mean()))
```

That's really low, so we're looking good here!

### Assumption 3: homoscedasticity

```python
sns.scatterplot(predictions, errors)
plt.title('Predictions vs. Residuals')
plt.xlabel('Predictions')
plt.ylabel('Residuals')
plt.axhline(y=0)
plt.show()
```

These seem pretty even until you're approaching 35, then we seem to be overestimating our predictions.

```python
from scipy.stats import bartlett
from scipy.stats import levene

bart_stats = bartlett(predictions, errors)
lev_stats = levene(predictions, errors)

print("Bartlett test statistic value is {} and p value is {}".format(bart_stats[0], bart_stats[1]))
print("Levene test statistic value is {} and p value is {}".format(lev_stats[0], lev_stats[1]))
```

This could be due to outliers in our features. Let's see...

```python
for column in X:
    sns.boxplot(X[column])
    plt.title('{}'.format(column))
    plt.show()
```

Looks like windspeed has some outliers on the right-side. Could be causing our heteroscedasticity. We could also be missing an important feature in our model

### Assumption 4: Muticollinearity

```python
sns.heatmap(X.corr(), annot=True, linewidth=0.5)
plt.show()
```

These seem very uncorrelated. So that's good news!

### Assumption 5: Uncorrelated error terms

```python
plt.plot(errors)
plt.show()
```

Ok, this doesn't look too bad either!

### Assumption 6: Errors uncorrelated with features

```python
for column in X:
    print('{}: {}'.format(column, stats.pearsonr(X[column], errors)[0]))
```

These seem pretty uncorrelated! All good news, eh? Let's inspect the distribution of our errors...

```python
sns.distplot(errors)
plt.show()
```

This matches our heterscedasticity model as well. Both seem normally distributed but have long tails to the right!
